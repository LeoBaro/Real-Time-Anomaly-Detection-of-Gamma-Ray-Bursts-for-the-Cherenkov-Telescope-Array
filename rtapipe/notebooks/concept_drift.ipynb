```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from rtapipe.lib.dataset.data_manager import DataManager
from rtapipe.lib.datasource.Photometry3 import SimulationParams
```

```python
def plot_timeseries(data):
    fig, axes = plt.subplots(2,2, figsize=(10,5))
    axes = axes.ravel()
    for i in range(len(axes)):
        for er in range(3):
            y = data[i, :, er]
            x = range(len(y))
            axes[i].plot(x,y, marker="o", markersize=2, linestyle='dashed', linewidth=0.5)
            axes[i].set_ylim(0, 4e-9)
    plt.show()
```
